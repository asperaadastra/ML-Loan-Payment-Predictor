```python
dummy = pd.get_dummies(df['purpose'])
dummy
```

```python
df_new = pd.concat([df,dummy], axis='columns')
```

```python
df_new = df_new.drop(['purpose','all_other'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= df_new.drop('not.fully.paid', axis=1)
y = df_new['not.fully.paid']



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predictions),'\n', classification_report(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rtree = RandomForestClassifier(n_estimators=100)
```

```python
rtree.fit(X_train,y_train)
```

```python
predictions = rtree.predict(X_test)
```

```python
print(confusion_matrix(y_test, predictions),'\n', classification_report(y_test, predictions))
```
